### Как создать датафрейм

```python
df = pd.DataFrame({'type': ['A', 'A', 'B', 'B'], 'value' : ['10', '14', '12', '23']})
df
```

```python
df = pd.read_csv('my_stat.csv')
df
```

```python
subset_1 = df[['V1', 'V3']].head(10)
subset_1
```

```python
subset_2 = df[['V2', 'V4']].drop([1, 5])
subset_2
```

```python
# В переменную subset_1 сохраните только те наблюдения, у которых значения переменной V1  строго больше 0, 
# и значение переменной V3  равняется 'A'
subset_1 = df.query("(V1 > 0) & (V3 == 'A')")
subset_1
```

```python
# В переменную  subset_2  сохраните только те наблюдения, у которых значения переменной V2  не равняются 10, 
# или значения переменной V4 больше или равно 1.
subset_2 = df.query("(V2 != 10) & (V4 >= 1)")
subset_2
```

```python
df['V5'] = df['V1'] + df['V4']
df['V6'] = np.log(df['V2'])
df
```

```python
df = df.rename(columns={'V1' : 'session_value', 'V2' : 'group', 'V3' : 'time', 'V4' : 'n_users'})
df
```

```python
# В переменной session_value замените все пропущенные значения на нули.
df = df.fillna(0)
df
```

```python
df['n_users'] = ['NaN' if i < 0 else i for i in df['n_users']]
```

```python
df
```

```python
m = df['n_users'].median(skipna=True)
```

```python
df['n_users'] = [m if str(i) == 'NaN' else i for i in df['n_users']]
```

```python
df
```

```python
df['session_value'] = df['session_value'].fillna(0)
```

```python
# Рассчитайте среднее значение переменной session_value для каждой группы (переменная group), 
# в получившемся dataframe  переменная group не должна превратиться в индекс. 
# Также переименуйте колонку со средним значением session_value в mean_session_value.

# Получившийся результат сохраните в dataframe с именем mean_session_value_data.

my_stat = pd.read_csv('my_stat.csv')
my_stat
```

```python
my_stat = my_stat.rename(columns = {'V1' : 'session_value', 'V2' : 'group', 'V3' : 'time', 'V4' : 'n_users'})
my_stat
```

```python
my_stat.groupby('group', as_index=False).agg({'session_value' : 'mean'})
```

```python
mean_session_value_data = my_stat.groupby('group', as_index=False).agg({'session_value' : 'mean'}).rename(columns = {'session_value' : 'mean_session_value'})
```
