### Фильтрация с помощью loc и передачи условия

#### Выберем всех людей с gender == female

```python
students.gender == 'female'                                                 # поняли, где female, где male
```

```python
students.loc[students.gender == 'female']                                   # передали loc массив из True и False
```

```python
students.loc[students.gender == 'female', ['gender', 'writing score']]      # доп. параметры
```

#### Выберем всех людей c gender == female со значением writing score > среднего

```python
mean_value = students['writing score'].mean()
```

```python
students.loc[(students.gender == 'female') & (students['writing score'] > mean_value)]
```

**ВАЖНО : операторы and or умеют работать только с булевскими значениями, а здесь у нас булевый вектор, поэтому исопльзуем побитовые операции и со ()**

#### Определим у какой доли студентов в колонке lunch указано free/reduced

```python
result_students = students.loc[(students.lunch == 'free/reduced')]
```

```python
len(result_students) / len(students)
```

### Фильтрация с помощью query

```python
# Переменуем столбцы нашего датафрейма, чтобы работать было удобнее
students = students.rename(columns = {'parental level of education' : 'parental_level_of_education', 
                                      'test preparation course' : 'test_preparation_course', 
                                      'math score' : 'math_score',
                                      'reading score' : 'reading_score',
                                      'writing score' : 'writing_score'})
```

```python
students
```

```python
students.query("writing_score > 74")                 # query - запрос, то есть какое-то выражение, переданное в ("")
```

```python
students.query("gender == 'female'")                   # обратить внимание на комбинацию кавычек для сравнения со строкой!
```

```python
students.query("gender == 'male' & writing_score > 64")   # И и ИЛИ
```

```python
value = 64
students.query("writing_score > @value")                   # использование переменных в выражении через @ !
```

### Через обращение []

```python
students[students['parental_level_of_education'].isin(["bachelor's degree", "master's degree"])]
```

```python
students[(students['math_score'] > 90) & (students['gender'] == 'female')]
```

```python
students[students['test_preparation_course'] != 'completed']
```

### Получить отдельные столбцы

* **Лапками**

```python
students[['math_score', 'writing_score']]                            # не пандовский способ
```

* **Через названия колонок**

```python
list(students)                                                                     # получить все названия колонок
```

```python
score_columns = [i for i in list(students) if 'score' in i]                        # сделать какую-то выборку по колонкам
```

```python
students[score_columns]
```

* **Пандосовские методы (с помощью filter)**

```python
students.filter(like='score')
```

students.filter(like='0', axis=0)                                          # Можно применять и для строк, AXIS !!!

Так же у этого метода есть следующие параметры:

    * items - отбирает записи, которые были переданы в качестве значения
    * regex - отбирает записи, подходящие к переданному паттерну
    
**!** Можно использовать только один из like, item, regex
