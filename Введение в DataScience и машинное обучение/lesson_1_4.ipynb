### Отбор отдельных частей таблицы   (методы *iloc* и *index + loc*)

```python
students.iloc[0:5, 0:3]                                   # Найти по номеру (integer location)
```

```python
students.iloc[[1, 4, 6], [3, -1, -2]]                       # По конкретным строкам и столбцам
```

**Создадим новую таблицу, в которой заменим порядковые номера строк на конкретные значения**

```python
students_with_names = students.iloc[0:5]
```

```python
students_with_names.index = ['Nastya', 'Marta', 'Olga', 'Serg', 'Gleb']
```

```python
students_with_names
```

```python
students_with_names.loc[['Olga', 'Gleb'], ['gender', 'lunch']]
```

### Одномерные массивы *Series* и многомерные *DataFrame*

```python
type(students_with_names.loc[['Olga', 'Gleb'], ['gender', 'lunch']])       # pandas.core.frame.DataFrame
```

```python
type(students_with_names.loc[:, ['gender']])                               # pandas.core.frame.DataFrame
```

```python
type(students_with_names.iloc[:, 0])                                       # pandas.core.series.Series
```

**Создание Series и объединение их в DataFrame**

```python
ser_1 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
```

```python
ser_2 = pd.Series([4, 5, 6], index=['a', 'b', 'c'])
```

```python
ser_1
```

```python
pd.DataFrame({'first_column' : ser_1, 'second_column' : ser_2})
```

**То есть DataFrame == объединенные колонки Series**

##### Если бы колонки не совпадали по индексам

```python
ser_3 = pd.Series([7, 8, 9, 10], index=['r', 'b', 'c', 'e'])
```

```python
pd.DataFrame({'first_column' : ser_1, 'second_column' : ser_2, 'third_column' : ser_3})
```

То есть мы видим, что оставшиеся значения заполняются NaN

##### Еще одна интересная особенность или как взять значения соответствующего столбца

```python
students_with_names['gender']                                   # Series (возьми то, что лежит под этим именем)
```

```python
students_with_names[['gender']]                                 # DataFrame (возьми первую колонку)
```

**!** В зависимости от типа изменяется и работа методов, например

```python
students_with_names['gender'].shape                             # (5,)
```

```python
students_with_names[['gender']].shape                           # (5, 1)
```
