```python
import pandas as pd
import numpy as np
```

```python
students = pd.read_csv('StudentsPerformance.csv')
```

```python
students
```

#### Посчитаем средние значения score в зависимости от пола студентов

```python
students.groupby('gender').mean()
```

```python
students.groupby('gender').aggregate({'math score' : 'mean', 'reading score' : 'min'})
```

Мы видим, что получился датафрейм, у которого индекстовая переменная - столбец, по которому мы группировали

Если нам это не удобно, то можно использовать параметр **as_index = False**

```python
students.groupby('lunch', as_index=False).aggregate({'math score' : 'min'})
```

```python
# Переименуем колонку, чтобы было понятнее, что там лежит
students.groupby('lunch', as_index=False).aggregate({'math score' : 'mean'}).rename(columns={'math score' : 'math_mean'})
```

#### Группировка по нескольким колонкам

```python
students.groupby(['gender', 'lunch']).mean()
```

```python
# При такой записи получаются интересные индексы
mean_scores= students.groupby(['gender', 'lunch']).mean()
mean_scores.index
```

```python
# Как работает индексация в этом случае
mean_scores.loc[('female', 'standard')]
```

```python
# Достать несколько 
mean_scores.loc[[('female', 'standard'), ('male', 'standard')]]
```

```python
# Получим количество уникальных значений в каждой из групп
students.groupby(['gender', 'lunch'])['math score'].nunique()                       # При чем это все еще Serial    !!!
```

#### Найдем топ-5 математиков-женщин и топ-5 математиков-мужчин

```python
# Отсортируем данные
sort = students.sort_values(['gender', 'math score'], ascending=False)         # ascending - по убыванию
```

```python
sort
```

```python
sort.groupby('gender').head()
```

### Создание новых колонок

```python
students['total score'] = students['math score'] + students['reading score'] + students['writing score']
```

```python
students
```

```python
# Способ, позволяющий делать более сложные штуки
students = students.assign(total_log_score = np.log(students['total score']))
```

```python
students
```

### Удалять колонки

```python
students.drop(['total_log_score', 'total score'], axis=1)
```
